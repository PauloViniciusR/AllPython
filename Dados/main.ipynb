```python
import matplotlib.pyplot as plt 

dias = ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado', 'Domingo']
vendas = [300, 450, 600, 355, 500, 700, 350]

plt.bar(dias, vendas, color=['blue', 'green', 'red', 'purple', 'orange', 'yellow', 'Darkred'])

plt.title('Vendas durante a semana')
plt.xlabel('Dias da semana')
plt.ylabel('Vendas da semana')

plt.show()

```
